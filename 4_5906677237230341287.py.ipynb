## Prediction de la natremie en fonction de la calcemie, de la chloremie et de la kaliemie


```python
dtc3 = dtc1.copy()
dtc3.drop(['GLUCOSE', 'MAGNESIUM', 'PHOSPHATE', 'LITHIUM', 'CREAT', 'UREA', 'gender', 'height', 'weight_avg'], 1, inplace=True)
dtc3.dropna(inplace=True)
```

```python
dtc3.head()
```

```python
X = dtc3[['CALCIUM', 'CHLORIDE', 'POTASSIUM']]
Y = dtc3['SODIUM']

display(Y.describe())
```

```python
X.info()
```

### Utilisation d'un modele de Regression Lineaire

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression, HuberRegressor
from sklearn import metrics

X_train, X_test, Y_train, Y_test = train_test_split (X, Y, train_size = 0.75, random_state=0)
```

```python
modelLIR=LinearRegression()
modelLIR.fit(X_train, Y_train)
```

```python
modelLIR.coef_
```

```python
Y_predictLIR= modelLIR.predict(X_test)
Y_pr_dfLIR = pd.DataFrame({'Y_test':Y_test, 'Y_predict': Y_predictLIR})
Y_pr_dfLIR
```

 ## REGRESSION LOGISTIQUE

```python

modelLOR=LogisticRegression()
modelLOR.fit(X_train, Y_train)
Y_predictLOR=modelLOR.predict(X_test)
Y_pr_dfLOR = pd.DataFrame({'Y_test':Y_test, 'Y_predictLOR': Y_predictLOR})
Y_pr_dfLOR
```

## HUBER REGRESSOR 

```python

modelHUR=HuberRegressor(epsilon=1.35)
modelHUR.fit(X_train, Y_train)
Y_predictHUR=modelHUR.predict(X_test)
Y_pr_dfHUR = pd.DataFrame({'Y_test':Y_test, 'Y_predict': Y_predictHUR})
Y_pr_dfHUR
```
