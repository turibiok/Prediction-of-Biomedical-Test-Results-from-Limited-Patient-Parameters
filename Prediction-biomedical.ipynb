## Prediction de la natremie en fonction de la calcemie, de la chloremie et de la kaliemie
### Utilisation d'un modele de regression lineaire

```python
dtc3 = dt.copy()
dtc3.drop(['charttime', 'subject_id', 'GLUCOSE', 'MAGNESIUM', 'PHOSPHATE', 'CREAT', 'UREA', 'gender', 'height', 'weight_avg'], 1, inplace=True)
dtc3.dropna(inplace=True)
```

```python
X = dtc3[['CALCIUM', 'CHLORIDE', 'POTASSIUM']]
Y = dtc3['SODIUM']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, train_size = 0.75, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train, Y_train)
```

```python
model.coef_
```

```python
Y_predict=model.predict(X_test)
Y_pr_df = pd.DataFrame({'Y_test':Y_test, 'Y_predict': Y_predict})
Y_pr_df
```
