# Preprocessing

```python
data95 = result_data.copy()
```

```python
data95.drop(['subject_id', 'chartime'], axis=1, inplace=True)
```

### Gestion des outliers

```python
def outlier(df):
    for i in df.select_dtypes('float'):
        #limite superieur
        upper_limit = df[i].quantile(0.95)
        #limite inferieur
        lower_limit = df[i].quantile(0.05)
        df[i] = df[i].clip(lower = lower_limit, upper =  upper_limit)
    return df
```

### Valeur manquantes

```python
def imputation(df):
    for i in df.select_dtypes('float'):
        df[i].fillna((df[i].median()), inplace=True)
    for a in df.select_dtypes('object'):
                     df[a] = df[a].fillna(value = 'M')
    return df
```

### Encodage de variable

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def encodage(df):
    labelencoder = LabelEncoder()
    df['gender'] = labelencoder.fit_transform(df['gender'])
    return df
```

### Train-Test

```python
from sklearn.model_selection import train_test_split
```

```python
trainset, testset = train_test_split(data95, test_size=0.2, random_state=0)
```

```python
trainset.shape, testset.shape
```

### Fonction de préprocessing

```python
def preprocessing(df) :
    df= outlier(df)
    df= imputation(df)
    df= encodage(df)
    X = df.drop('PHOSPHATE', axis=1)
    y = df['PHOSPHATE']
    print(X.shape, y.shape)
    
    return X, y
```

```python
X_train, y_train = preprocessing(trainset)
```

```python
X_test, y_test = preprocessing(testset)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
preprocessor = make_pipeline(SelectKBest(f_regression, k =10), MinMaxScaler())
```

```python
LRegression = make_pipeline(preprocessor, LinearRegression())
DecisionTree = make_pipeline(preprocessor, DecisionTreeRegressor())
RandomForest= make_pipeline(preprocessor, RandomForestRegressor())
GradientBoosting = make_pipeline(preprocessor, GradientBoostingRegressor())
Xgboost = make_pipeline(preprocessor, XGBRegressor()) 
```

```python
def evaluation(model):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    model_rmse = np.sqrt(mean_squared_error(pred, y_test))
    model_rmse_train = np.sqrt(mean_squared_error(model.predict(X_train), y_train))
    print('train_error :', model_rmse_train)
    print(print('test_error :', model_rmse))
```

```python
dict_model = {
    'LRegression':LRegression,
    'DecisionTree':DecisionTree,
    'RandomForest':RandomForest,
    'GradientBoosting':GradientBoosting,
    'Xgboost':Xgboost   
}
```

```python
for name, model in dict_model.items():
    print('--------------------',name,'----------------------')
    evaluation(model)   
```

### Optimisation 

```python
params_distrib = {
    'randomforestregressor__n_estimators':[100,180,200, 400, 600, 800, 1000],
    'randomforestregressor__max_features':[2,4,8,10],
    'randomforestregressor__max_depth':[10, 20, 30, 40, 50, 60, 70, 80]
}

reg_search = RandomizedSearchCV(RandomForest, param_distributions=params_distrib, n_iter=10, cv=5, 
                               scoring='neg_mean_squared_error', random_state=42)

reg_search.fit(X_train, y_train)


print(reg_search.best_params_)
```
